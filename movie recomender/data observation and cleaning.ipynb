```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('F:/M 20.02.2023/downloads/archive (5)/bollywood_data_set.csv')
```

```python
# Drop rows with missing release year
data = data.dropna(subset=['year_of_release'])
```

```python
data.info()
```

```python
# Standardize text columns (e.g., movie_name, director, actors)
data['movie_name'] = data['movie_name'].str.strip().str.title()
data['director'] = data['director'].str.strip().str.title()
data['actors'] = data['actors'].str.strip().str.title()  # Optional cleaning of actors' names
```

```python
data['runtime'] = data['runtime'].str.replace(' min', '', regex=False)

# Step 2: Remove any remaining non-numeric characters
data['runtime'] = pd.to_numeric(data['runtime'], errors='coerce')

# Step 3: Drop rows with invalid 'runtime' (if any non-numeric value was present)
data = data.dropna(subset=['runtime'])

# Convert runtime to integers
data['runtime'] = data['runtime'].astype(int)
```

```python
# Clean the 'votes' column (remove commas and convert to integers)
data['no_of_votes'] = data['no_of_votes'].str.replace(',', '', regex=False)
data['no_of_votes'] = pd.to_numeric(data['no_of_votes'], errors='coerce')
data = data.dropna(subset=['no_of_votes'])  # Drop rows where votes are invalid
data['no_of_votes'] = data['no_of_votes'].astype(int)
```
